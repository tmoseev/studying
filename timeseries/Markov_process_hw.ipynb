```python
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
list_tom = ['tom_1.txt', 'tom_2.txt', 'tom_3.txt', 'tom_4.txt']
wp = ''
```

```python
for elem in list_tom:
    with open(elem, 'r', encoding='utf-8') as tom:
        temp = tom.read().lower()
        wp += temp
wp = re.sub('[^ёа-я]','', wp)        
```

```python
russian = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
alphabet = {l: i for i, l in enumerate(russian)}
```

```python
prob = np.zeros((len(russian), len(russian)))
```

```python
for cp, cn in zip(wp[:-1], wp[1:]):
    prob[alphabet[cp], alphabet[cn]] += 1

prob = (prob / prob.sum(axis=1)[:, None])
```

```python
df = pd.DataFrame(prob, index = (letter for letter in russian), columns = (letter for letter in russian))
df
```

Из матрицы вероятностей видно, что после буквы 'э' пркатически во всех случаях идет буква 'т' (из очевидных примеров скорее всего слово 'это' и его различные вариации). Кроме этого часто можно встретить буквы 'ъ' перед 'е'. Это вероятнее всего связано со старорусской стилистикой.

```python
plt.figure(figsize=(20,14))
sns.heatmap(df)
plt.show()
```

```python
#Посчитайте вероятность встретить строку из 4-х символов, которая начинается п, а кончается на р
start = alphabet['п']
end = alphabet['р']

start_prob = 1. / len(russian)

for letter1 in russian:
    for letter2 in russian:
        proba = start_prob * prob[start, alphabet[letter1]] * prob[alphabet[letter1], alphabet[letter2]] * prob[alphabet[letter1], end]
        #оставим все вероятности больше >0.00001, поскольку общий список слишком большой
        if proba > 0.00001:
            print(''.join(['п', letter1, letter2, 'р']), proba)
```

```python
#Используя матрицу переходных вероятностей сгенерируйте слово, начинающееся с буквы “Б” и имеющее длину 8-мь символов
s = ['б']
start = np.random.choice(list(russian))
s.append(start)
length = 6
for i in range(length):
    index = alphabet[s[i]]
    next_char = np.random.choice(list(russian), p = prob[index])
    s.append(next_char)

print (''.join(s))
```

```python
dowjones_closing = pd.read_csv("Series/weekly-closings-of-the-dowjones-.csv")
dowjones = dowjones_closing['Close']
```

```python
#исходный начальный график
plt.style.use('bmh')
plt.figure(figsize = (16,6))
plt.plot(dowjones, label = 'Dowjones')
plt.legend()
plt.show()
```

```python
from hmmlearn.hmm import GaussianHMM
```

```python
dowjones_ = dowjones.values.reshape(-1,1)
```

```python
for n in range(1, 15):
    model = GaussianHMM(n_components=n,
                        covariance_type="diag",
                        n_iter=1000)
    model.fit(dowjones_)
    print (n, model.score(dowjones_))
```

Выберем 9 компонент

```python
model = GaussianHMM(n_components=9,
                        covariance_type="diag",
                        n_iter=1000)
model.fit(dowjones_)
labels = model.predict(dowjones_)
means = np.zeros_like(dowjones_)
for i in range(model.n_components):
    means[labels == i] = model.means_[i]
    
plt.figure(figsize=(12, 6))
plt.plot(dowjones_)
plt.plot(means, linewidth=3)
```
