```python
#задание1
def sourse_type(line):
    if line['traffic_source'] == 'yandex' or line['traffic_source'] == 'google':
        return 'organic'
    elif line['traffic_source'] == 'paid' and line['region'] == 'Russia':
        return 'ad'
    elif line['traffic_source'] == 'paid' and not line['region'] == 'Russia':
        return 'other'
    else:
        return line['traffic_source']
    
log['sourse_type'] = log.apply(sourse_type, axis = 1)
log
```

```python
ratings = pd.read_csv('C:\\jupyter\\pandas_numpy\\ml-latest-small\\ratings.csv')

rating_100 = ratings.groupby('userId').count()
user_rating_100 = rating_100[rating_100['rating'] >100]['rating'].reset_index()

max_time = ratings.groupby('userId').max()['timestamp'].reset_index()
max_time.rename(columns={'timestamp':'timestamp_max'}, inplace = True)

min_time = ratings.groupby('userId').min()['timestamp'].reset_index()
min_time.rename(columns={'timestamp':'timestamp_min'}, inplace = True)

total = user_rating_100.merge(max_time, on = 'userId').merge(min_time, on = 'userId')
total['lifetime'] = total.apply(lambda x: x['timestamp_max']-x['timestamp_min'], axis = 1)
total[['userId', 'rating', 'lifetime']].head(10)
```
