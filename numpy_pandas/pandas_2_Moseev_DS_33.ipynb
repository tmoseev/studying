```python
import pandas as pd
```

```python
#Задание1
ratings = pd.read_csv('C:\\jupyter\pandas_numpy\\ml-latest-small\\ratings.csv')
```

```python
def classification_of_films(movie_grade):
    if movie_grade < 2.0:
        return 'Низкий рейтинг'
    elif movie_grade < 4.0:
        return 'Средний рейтинг'
    else:
        return 'Высокий рейтинг'
```

```python
classification = ratings.groupby('movieId').mean()
classification['classification'] = classification['rating'].apply(classification_of_films)
classification.reset_index()[['movieId','rating', 'classification']]
```

```python
#задание2
keywords = pd.read_csv('C:\\jupyter\pandas_numpy\\ml-latest-small\\keywords.csv')
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def geo(words):
    keyword = words.split(' ')
    for key, value in geo_data.items():
        for elem in keyword:
            if elem in value:
                return key
    return 'undefined'
```

```python
keywords['location'] = keywords['keyword'].apply(geo)
keywords
```

```python
#задание3
ratings = pd.read_csv('C:\\jupyter\pandas_numpy\\ml-latest-small\\ratings.csv')
movies = pd.read_csv('C:\\jupyter\pandas_numpy\\ml-latest-small\\movies.csv')
general = ratings.merge(movies, how = 'left')
film_ratings = general.groupby('title').mean().reset_index()[['title', 'rating']]
years = [x for x in range(1950,2010)]
```

```python
def production_year(dataframe):
    for year in years:
        if str(year) in dataframe:
            return year
    return '1900'
```

```python
film_ratings['year'] = film_ratings['title'].apply(production_year)
film_ratings.sort_values('rating', ascending=False)
```
