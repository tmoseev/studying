```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

#Пользователь по команде "p" может узнать владельца документа по его номеру
def p_get_owner_by_number(array_of_documents):
    doc_number = input('Введите номер документа: ')
    for doc in array_of_documents:
        if doc_number == doc['number']:            
            return f"Владелец документа: {doc['name']}"
    else:
            return f"Документ не найден"
        
#Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится        
def s_get_shelf_by_number(array_of_directories, doc_number):
    find_result = 'Документ не найден в базе'
    for k in array_of_directories:
        if doc_number in array_of_directories[k]:
            find_result = 'Документ хранится на полке:'
            return f"{find_result} {k}"
    return find_result
        
#Пользователь по команде "l" может увидеть полную информацию по всем документам
def l_get_all_information(array_of_documents):
    print('Текущий список документов:')
    for elem in array_of_documents:
         print((f"№{elem['number']}, тип: {elem['type']}, владелец: {elem['name']}, \
{s_get_shelf_by_number(directories, doc_number = elem['number'])}"))
    return

#Пользователь по команде "ads" может добавить новую полку
def ads_add_new_shelf(array_of_directories):
    new_shelf = input('Введите номер полки: ')
    if new_shelf not in array_of_directories.keys():
        added_new_shelf = {new_shelf:[]}
        array_of_directories.update(added_new_shelf)
        return 'Полка добавлена. Текущий перечень полок:' + ', '.join(array_of_directories.keys())
    else:
        return 'Такая полка уже существует. Текущий перечень полок:' + ', '.join(array_of_directories.keys())

#Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
def ds_remove_shelf(array_of_directories):
    remove_shelf = input('Введите номер полки: ')
    if remove_shelf not in array_of_directories.keys():
        print(f"Такой полки не существует. Текущий перечень полок: {array_of_directories.keys()}")
    elif remove_shelf in array_of_directories.keys() \
        and (array_of_directories.get(remove_shelf) == None or array_of_directories.get(remove_shelf) == []):
        del array_of_directories[remove_shelf]
        return 'Полка удалена. Текущий перечень полок:' + ', '.join(array_of_directories.keys())
    else:
        return 'На полке есть документы, удалите их перед удалением полки.\
Текущий перечень полок:' + ', '.join(array_of_directories.keys())

#Пользователь по команде "ad" может добавить новый документ в данные
def ad_add_new_data(array_of_documents, array_of_directories):
    new_number = input('Введите номер документа: ')
    new_type =  input('Введите тип документа: ')
    new_owner = input('Введите владельца документа: ')
    new_shelf = input('Введите полку для хранения: ')
    temp_dict = {'type': new_type, 'number': new_number, 'name': new_owner}
    for element in array_of_directories.items():
        if new_shelf == element[0]:
            array_of_documents.append(temp_dict)
            list_documents = element[1]
            list_documents.append(new_number)
            array_of_directories.setdefault(element[0], list_documents)
        elif new_shelf not in array_of_directories.keys():
            print('Сначала добавьте полку командой ads')
            break
    l_get_all_information(documents)
    
def main(array_of_documents, array_of_directories):
    while True:
        command = input('Введите команду: ')
        if command == 'p':
            print(p_get_owner_by_number(documents))  
        elif command == 's':
            print(s_get_shelf_by_number(directories, doc_number = input('Введите номер документа: ')))
        elif command == 'l':
            l_get_all_information(documents)
        elif command == 'ads':
            print(ads_add_new_shelf(directories)) 
        elif command == 'ds':
            print(ds_remove_shelf(directories)) 
        elif command == 'ad':
            ad_add_new_data(documents,directories)     
        elif command == 'q':
            print('Программа завершена')
            break   
            
main(documents, directories)
```

>при исполнении команды l мы видим лишний None в конце. Так происходит потому, что при вызове функции внутри print на экран выводится то, что она возвращает. Но эта функция у вас ничего не возвращает, а вы внутри нее уже используете print для вывода нужной информации. Поэтому в данном случае можно вызывать ее не внутри print, а просто саму по себе;


Код был пересмотрен, лишние print'ы были убраны. Теперь лишних None не выводится.
>при добавлении новой полки она у вас инициализируется None (а надо пустым списком), что приведет к дальнейшей невозможности корректно с ней работать;

В функции ads инициализация полки была изменена.
>при добавлении документа на несуществующую полку он у вас добавляется в documents, а в directories – нет, что приводит к несогласованности данных. Лучше тогда вообще ничего никуда не добавлять и выводить предупреждение, либо сразу эту новую полку создавать;

В функции ad добавление полки было перемещено в блок проверки наличия полки. Если такая полка существует, то и все остальные данные добавляются. Если не существует, то документы не добавляются не несущессвующую полку.
>финальный блок кода с обработкой пользовательского ввода лучше поместить в отдельную функцию, чтобы в глобальном контексте не оставалось ничего лишнего.

Финальный блок кода был помещен в функцию main()
