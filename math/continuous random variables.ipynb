```python
import matplotlib.pyplot as plt
import scipy.stats as sts
import scipy.integrate as integrate
import numpy as np
import math
```

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
parametrs_normal = [[1, 0.5],[3, 1.5], [5,5], [7, 9], [8,6]]
x = np.linspace(0,10,100)
```

```python
for mu, sigma in parametrs_normal:
    norm_rv = sts.norm(loc = mu, scale = sigma)
    l = f'loc = {mu}, scale = {sigma}'
    plt.plot(x, norm_rv.cdf(x), label = l)
plt.legend()
plt.show()
```

```python
for mu, sigma in parametrs_normal:
    norm_rv = sts.norm(loc = mu, scale = sigma)
    l = f'loc = {mu}, scale = {sigma}'
    plt.plot(x, norm_rv.pdf(x), label = l)
plt.legend()
plt.show()
```

```python
parametrs_expoten = [0.1, 0.5, 3, 6, 15]
```

```python
for lam in parametrs_expoten:
    expon_rv = sts.expon(lam)
    l = f'lambda = {lam}'
    plt.plot(x, expon_rv.cdf(x), label = l)
plt.legend()
plt.show()
```

```python
for lam in parametrs_expoten:
    expon_rv = sts.expon(lam)
    l = f'lambda = {lam}'
    plt.plot(x, expon_rv.pdf(x), label = l)
plt.legend()
plt.show()
```

```python
for lam in parametrs_expoten:
    expon_rv = sts.expon(lam)
    l = f'lambda = {lam}'
    plt.plot(x, expon_rv.sf(x), label = l)
plt.legend()
plt.show()
```

```python
parametrs_student = [0.05, 0.5, 4, 8, 22]
```

```python
for n in parametrs_student:
    t_rv = sts.t(n)
    l = f'n = {n}'
    plt.plot(x, t_rv.cdf(x), label = l)
plt.legend()
plt.show()
```

```python
for n in parametrs_student:
    t_rv = sts.t(n)
    l = f'n = {n}'
    plt.plot(x, t_rv.pdf(x), label = l)
plt.legend()
plt.show()
```

```python
for n in parametrs_student:
    t_rv = sts.t(n)
    l = f'n = {n}'
    plt.plot(x, t_rv.sf(x), label = l)
plt.legend()
plt.show()
```
