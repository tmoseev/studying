```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x + 3)
```

```python
#первая производная в точке х = 1
derivative(f,1.0)
```

```python
#вторая производная в точке х = 1 
derivative(f,1.0,n=2)
```

```python
def g(y):
    return (np.sin(2*y+1))**5
```

```python
#график первой производной
y = np.linspace(-5,5)
gy = g(y)
f1x=[derivative(g,number) for number in y]
plt.plot(y,f1x, label = 'first derivative')
plt.legend()
plt.show()
```

```python
#график второй производной
f2x=[derivative(g,number,n=2) for number in y]
plt.plot(y,f2x, label = 'second derivative')
plt.legend()
plt.show()
```
